```python
#function that calculate the dot product manualy
a = [2,6]
b = [3,7]
c = []

def dot_product(v1, v2):
    for n1, n2 in zip(a, b):
        c.append(n1 * n2)
    
    for i in range(len(c)):
        for j in range(i+1, len(c)):
            output = c[i]+c[j]
    return output
```

A one hot encoding allows the representation of categorical data to be more expressive. Many machine learning algorithms cannot work with categorical data directly. The categories must be converted into numbers. This is required for both input and output variables that are categorical.
One hot encoding is a technique which converts categorical variables to numerical in an interpretable format.

```python
import pandas as pd
import numpy as np

belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
belize_df.head()
```
